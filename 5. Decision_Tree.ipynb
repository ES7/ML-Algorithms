# Decision Tree
A decision tree is a **non-parametric supervised learning** algorithm for classification and regression tasks. It has a **hierarchical tree** structure consisting of a root node, branches, internal nodes, and leaf nodes.<br>
A decision tree is a hierarchical model used in decision support that depicts decisions and their potential outcomes, incorporating chance events, resource expenses, and utility. This algorithmic model utilizes conditional control statements and is non-parametric, supervised learning, useful for both classification and regression tasks. The tree structure is comprised of a root node, branches, internal nodes, and leaf nodes, forming a hierarchical, tree-like structure.<br>
It is a tool that has applications spanning several different areas. The name itself suggests that it uses a flowchart like a tree structure to show the predictions that result from a series of feature-based splits. It starts with a root node and ends with a decision made by leaves.<br>

## Decision Tree Terminologies
* **Root Node:** The initial node at the beginning of a decision tree, where the entire population or dataset starts dividing based on various features or conditions.<br>
* **Decision Nodes:** Nodes resulting from the splitting of root nodes are known as decision nodes. These nodes represent intermediate decisions or conditions within the tree.<br>
* **Leaf Nodes:** Nodes where further splitting is not possible, often indicating the final classification or outcome. Leaf nodes are also referred to as terminal nodes.<br>
* **Sub-Tree:** Similar to a subsection of a graph being called a sub-graph, a sub-section of a decision tree is referred to as a sub-tree. It represents a specific portion of the decision tree.<br>
* **Pruning:** The process of removing or cutting down specific nodes in a decision tree to prevent overfitting and simplify the model.<br>
* **Branch / Sub-Tree:** A subsection of the entire decision tree is referred to as a branch or sub-tree. It represents a specific path of decisions and outcomes within the tree.<br>
* **Parent and Child Node:** In a decision tree, a node that is divided into sub-nodes is known as a parent node, and the sub-nodes emerging from it are referred to as child nodes. The parent node represents a decision or condition, while the child nodes represent the potential outcomes or further decisions based on that condition.<br>

## How Decision Tree Algorithms Works?
**1. Starting at the Root:** The algorithm begins at the top, called the "root node", representing the entire dataset.<br>
**2. Asking the Best Questions:** It looks for the most important feature or question that splits the data into the most distinct groups. This is like asking a question at a fork in the tree.<br>
**3. Branching Out:** Based on the answer to that question, it divides the data into smaller subsets, creating new branches. Each branch represents a possible route through the tree.<br>
**4. Repeating the Process:** The algorithm continues asking questions and splitting the data at each branch until it reaches the final "leaf nodes", representing the predicted outcomes or classifications.<br>

## Assumptions in Decision Tree
Several assumptions are made to build effective models when creating decision trees. These assumptions help guide the tree's construction and impact its performance. Here are some common assumptions and considerations when creating decision trees:<br>

#### 1. Binary Splits
Decision trees typically make binary splits, meaning each node divides the data into two subsets based on a single feature or condition. This assumes that each decision can be represented as a binary choice.

#### 2. Recursive Partitioning
Decision trees use a recursive partitioning process, where each node is divided into child nodes, and this process continues until a **stopping criterion** is met. This assumes that data can be effectively subdivided into smaller, more manageable subsets.

#### 3. Feature Independence
Decision trees often assume that the features used for splitting nodes are independent. In practice, feature independence may not hold, but decision trees can still perform well if features are correlated.

#### 4. Homogeneity
Decision trees aim to create homogeneous subgroups in each node, meaning that the samples within a node are as similar as possible regarding the target variable. This assumption helps in achieving clear decision boundaries.

#### 5. Top-Down Greedy Approach
Decision trees are constructed using a top-down, greedy approach, where each split is chosen to maximize **information gain** or **minimize impurity** at the current node. This may not always result in the **globally optimal tree**.

#### 6. Categorical and Numerical Features
Decision trees can handle both categorical and numerical features. However, they may require different splitting strategies for each type.

#### 7. Overfitting
Decision trees are prone to overfitting when they capture **noise** in the data. Pruning and setting appropriate stopping criteria are used to address this assumption.

#### 8. Impurity Measures
Decision trees use impurity measures such as **Gini impurity** or **entropy** to evaluate how well a split separates classes. The choice of impurity measure can impact tree construction.

#### 9. No Missing Values
Decision trees assume that there are no missing values in the dataset or that missing values have been appropriately handled through imputation or other methods.

#### 10. Equal Importance of Features
Decision trees may assume equal importance for all features unless feature scaling or weighting is applied to emphasize certain features.

#### 11. No Outliers
Decision trees are sensitive to outliers, and extreme values can influence their construction. Preprocessing or robust methods may be needed to handle outliers effectively.

#### 12. Sensitivity to Sample Size
Small datasets may lead to overfitting, and large datasets may result in overly complex trees. The sample size and tree depth should be balanced.

## When to Stop Splitting?
When do we stop growing our Decision tree? Usually, real-world datasets have a large number of features, which will result in a large number of splits, which in turn gives a huge tree. Such trees take time to build and can lead to overfitting. That means the tree will give very good accuracy on the training dataset but will give bad accuracy in test data.

There are many ways to tackle this problem through hyperparameter tuning. We can set the **maximum depth** of our decision tree using **themax_depth** parameter. The more the value of **max_depth**, the more complex your tree will be. The training error will off-course decrease if we increase the max_depth value but when our test data comes into the picture, we will get a very bad accuracy. Hence our need a value that will not overfit as well as underfit our data and for this, we can use **GridSearchCV**.

Another way is to set the minimum number of samples for each spilt. It is denoted by **min_samples_split**. Here we specify the minimum number of samples required to do a spilt. For example, we can use a minimum of 10 samples to reach a decision. That means if a node has less than 10 samples then using this parameter, we can stop the further splitting of this node and make it a leaf node.

There are more hyperparameters such as :<br>
* **min_samples_leaf –** represents the minimum number of samples required to be in the leaf node. The more you increase the number, the more is the possibility of overfitting.<br>
* **max_features –** it helps us decide what number of features to consider when looking for the best split.<br>

#### Pruning
Pruning is another method that can help us avoid overfitting. It helps in improving the performance of the Decision tree by cutting the nodes or sub-nodes which are not significant. Additionally, it removes the branches which have very low importance.

There are mainly 2 ways for pruning:<br>
* **Pre-pruning –** we can stop growing the tree earlier, which means we can prune/remove/cut a node if it has low importance while growing the tree.<br>
* **Post-pruning –** once our tree is built to its depth, we can start pruning the nodes based on their significance.

```python
class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None,*,value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value
        
    def is_leaf_node(self):
        return self.value is not None
```

The class **`Node`** represents a node in a decision tree.<br>
The **`__init__()`** functoin is the constructor method for the **Node** class. It initalizes the attributes of the node.<br>
* The **feature** attribute stores the index of the feature that is used for splitting at this node. If the node is a leaf node then this attribute is **None**.
* The **threshold** attribute stores the threshold value for the feature used for splitting at this node. If the node is a leaf node, this attribute is None.
* The **left** and **right** attributes store references to the left and right child nodes of the current node. These child nodes represent the branches that the decision tree can take based on the feature's value being less than or equal to the threshold (left) or greater than the threshold (right). If the node is a leaf node, both left and right attributes are None.
* The **value** attribute stores the predicted value for the leaf node. For regression tasks, this could be the mean of the target values in the leaf node's training instances. For classification tasks, this could be the majority class label in the leaf node's training instances.

The **`is_leaf_node`** method is used to check if the current node is a leaf node. It returns True if the node is a leaf node (i.e., if value is not None), and False otherwise.

## Decision Tree for Classification Task

```python
class DecisionTreeClassification:
    def __init__(self, min_samples_split=2, max_depth=100, n_features=None):
        self.min_samples_split=min_samples_split
        self.max_depth=max_depth
        self.n_features=n_features
        self.root=None

    def fit(self, X, y):
        self.n_features = X.shape[1] if not self.n_features else min(X.shape[1],self.n_features)
        self.root = self._grow_tree(X, y)

    def _grow_tree(self, X, y, depth=0):
        n_samples, n_feats = X.shape
        n_labels = len(np.unique(y))

        # check the stopping criteria
        if (depth>=self.max_depth or n_labels==1 or n_samples<self.min_samples_split):
            leaf_value = self._most_common_label(y)
            return Node(value=leaf_value)

        feat_idxs = np.random.choice(n_feats, self.n_features, replace=False)

        # find the best split
        best_feature, best_thresh = self._best_split(X, y, feat_idxs)

        # create child nodes
        left_idxs, right_idxs = self._split(X[:, best_feature], best_thresh)
        left = self._grow_tree(X[left_idxs, :], y[left_idxs], depth+1)
        right = self._grow_tree(X[right_idxs, :], y[right_idxs], depth+1)
        return Node(best_feature, best_thresh, left, right)


    def _best_split(self, X, y, feat_idxs):
        best_gain = -1
        split_idx, split_threshold = None, None

        for feat_idx in feat_idxs:
            X_column = X[:, feat_idx]
            thresholds = np.unique(X_column)

            for thr in thresholds:
                # calculate the information gain
                gain = self._information_gain(y, X_column, thr)

                if gain > best_gain:
                    best_gain = gain
                    split_idx = feat_idx
                    split_threshold = thr

        return split_idx, split_threshold


    def _information_gain(self, y, X_column, threshold):
        # parent entropy
        parent_entropy = self._entropy(y)

        # create children
        left_idxs, right_idxs = self._split(X_column, threshold)

        if len(left_idxs) == 0 or len(right_idxs) == 0:
            return 0
        
        # calculate the weighted avg. entropy of children
        n = len(y)
        n_l, n_r = len(left_idxs), len(right_idxs)
        e_l, e_r = self._entropy(y[left_idxs]), self._entropy(y[right_idxs])
        child_entropy = (n_l/n) * e_l + (n_r/n) * e_r

        # calculate the IG
        information_gain = parent_entropy - child_entropy
        return information_gain

    def _split(self, X_column, split_thresh):
        left_idxs = np.argwhere(X_column <= split_thresh).flatten()
        right_idxs = np.argwhere(X_column > split_thresh).flatten()
        return left_idxs, right_idxs

    def _entropy(self, y):
        hist = np.bincount(y)
        ps = hist / len(y)
        return -np.sum([p * np.log(p) for p in ps if p>0])


    def _most_common_label(self, y):
        counter = Counter(y)
        value = counter.most_common(1)[0][0]
        return value

    def predict(self, X):
        return np.array([self._traverse_tree(x, self.root) for x in X])

    def _traverse_tree(self, x, node):
        if node.is_leaf_node():
            return node.value

        if x[node.feature] <= node.threshold:
            return self._traverse_tree(x, node.left)
        return self._traverse_tree(x, node.right)
```

This class represents a **Decision Tree** model.<br>
The **`__init__()`** is a constructor method for the class which initializes the attributes of the decision tree model.<br>
* **min_samples_split** attribute specifies the minimum number of samples required to split an internal node. If a node has fewer samples than min_samples_split, it will not be split, and it will become a leaf node.
* **max_depth** attribute specifies the maximum depth of the decision tree. If set to None, the tree will grow until all leaves are pure or until all leaves contain less than min_samples_split samples.
* **n_features** attribute specifies the number of features to consider when looking for the best split. If set to None, all features will be considered.
* **root** attribute stores the root node of the decision tree. Initially, it is set to None because the tree is empty.

The **`fit()`** function is used to train the model.<br>
* **X** is the input features matrix, where each row represetns a sample and each column represents a feature.
* **y** in the target variable vector, which contains the labels corresponding to each sample in 'X'.
* **n_features** is set to the number of features in the input data as **X.shape[1]** if it is 'None' and if it is 'not None' then it is set to the minimum nomber of features in 'X' and the specified value. This step is done to hange the case where 'n_features' is explicitly set by the user to limit the number of features considered for splitting.
* **root** calls the **_grow_tree** method to recursively grow the decision tree using the input data X and target variable y. The _grow_tree method is responsible for building the tree structure by recursively splitting nodes based on the input data and target variable.<br>

The **`_grow_tree()`** function is a recursive function used to build the decision tree structure.<br>
* X** is the input features matrix, where each row represents a sample and each column represents a feature.
* **y** is the target variable vector, which contains the labels corresponding to each sample in X.
* **depth** is the current depth of the node in the tree.
**n_samples, n_feats**, this line of code retrieves the number of samples (n_samples) and the number of features (n_feats) in the input data X.
* **n_labels** This line of code calculates the number of unique labels in the target variable y. This information is useful for determining when to stop splitting nodes further. If n_labels is 1 (i.e., all samples in the node have the same label), the node becomes a leaf node and no further splitting is needed.<br>

The **_grow_tree()** works as follows:<br>
* Check if the stopping criteria for splitting a node are met. This could include conditions such as reaching the maximum depth (self.max_depth), having fewer samples than min_samples_split, or having all samples belong to the same class.
* If the stopping criteria are met, create a leaf node with the majority class label as the predicted value and return the node.
* If the stopping criteria are not met, find the best split for the current node. This involves iterating over all features and their possible thresholds to find the split that maximizes a certain criterion (e.g., Gini impurity or information gain).
* Split the node into two child nodes based on the best split found in the previous step.
Recursively call _grow_tree on the left and right child nodes to continue splitting the tree.

* Stopping criteria: The function first checks if the stopping criteria for splitting a node are met. These criteria include:<br>
    * **`depth >= self.max_depth`**: If the current depth of the node is equal to or exceeds the maximum depth specified for the tree (self.max_depth), stop splitting and create a leaf node.
    * **`n_labels == 1`**: If all samples in the node belong to the same class (i.e., there is only one unique label), stop splitting and create a leaf node.
    * **`n_samples < self.min_samples_split`** If the number of samples in the node is less than the minimum number of samples required to split a node (self.min_samples_split), stop splitting and create a leaf node.

    * If the stopping criteria are met, the function calculates the most common label in the target variable y and creates a leaf node with this label as the predicted value.

    * Random feature selection: The function randomly selects a subset of features (self.n_features) without replacement using np.random.choice. This is useful for Random Forests and helps to introduce randomness and reduce overfitting.

* Then we find the best feature and threshold value for splitting the node based on the selected subset of features. The _best_split method is responsible for finding the best split based on a certain criterion (e.g., Gini impurity or information gain).

* Then we split the samples in the node into two groups based on the best split found in the previous step. The _split method is responsible for dividing the samples into left and right groups based on a threshold value.

* Recursive calls: The function recursively calls _grow_tree on the left and right groups of samples to continue building the tree structure.

* Finally, the function returns a node representing the current split, with references to the left and right child nodes. If the current node is a leaf node, left and right will be None.

The **`_best_split()`** function is responsible for finding the best feature and threshold value to split a node based on a certain criterion (e.g., information gain or gini impurity).<br>
* **feat_idxs** is a subset of feature indices to consider for splitting.
* **best_gain = -1** stores the best information gain found so far. It is initialized to -1 to ensure that any positive gain found during the iteration will be considered as an improvement.
* **split_idx** and **split_threshold** variables store the index and threshold of the feature  for the best split found so far. They are initialize to **None**.
* The we extracts the column of the feature being considered for splitting and finds unique values in the feature column, which will be used as potential split thresholds.
* **Information gain calculation:** For each unique value in the feature column, the method calculates the information gain using the **_information_gain** method. Information gain is a measure of the improvement in the model's performance achieved by splitting the node based on a particular feature and threshold.
* **Update best split:** If the calculated gain is greater than the current best gain, the method updates **best_gain**, **split_idx**, and **split_threshold** to the values corresponding to the current feature and threshold and return the bebst split value.

The **`_information_gain()`** function is used to calculate the IG achieved by splitting a node based on a particular feature and threshold.<br>
* **parent_entropy** calculates the entropy of the parent node before the split. Entropy is a measure of impurity or disorder in a set of labels. The _entropy method is used to calculate the entropy of the target variable y.
* Then we split the samples into two groups based on the threshold value. The _split method is responsible for dividing the samples into left and right groups based on a threshold value.
* The check if either the left or right group is empty after the split. If either group is empty, it means that the split did not partition the data, and the information gain is 0.
* **n_l** and **n_r** stores the number of samples in the left and right groups, respectively, and **e_l** and **e_r** store the entropy of the left and right groups, respectively. The _entropy method is used to calculate the entropy of each group.
* **child_entropy** calculates the weighted average entropy of the two child nodes. The weight of each child node is the proportion of samples in that node relative to the total number of samples.
* Then we calculate the information gain, which is the difference between the entropy of the parent node and the weighted average entropy of the child nodes. Higher information gain indicates that the split is more informative. The function returns the calculated information gain.

The **`_split()`** function  is used to split a feature column into two groups based on a given threshold value.<br>
* **left_idxs** finds the indices of the elements in the feature column that are less than or equal to the split_thresh. The **np.argwhere** function returns the indices of elements that satisfy the condition as a 2D array. The **flatten()** method is used to convert this 2D array into a 1D array of indices.
* **right_idxs** finds the indices of the elements in the feature column that are greater than the split_thresh. Similar to the previous line, it uses np.argwhere and flatten() to convert the 2D array of indices into a 1D array.
* The function returns left_idxs and right_idxs, which represent the indices of the samples that belong to the left and right groups, respectively, after the split.

The **`entropy()`** function calculate the entropy of a set of labels.<br>
* **hist = np.bincount(y)** calculates the histogram of the labels y. It counts the occurrences of each label in the target variable and returns an array where the value at index i is the count of occurrences of label i. Then we divide this with the total number of samples and then return the entropy.

The **`_most_common_label()`** function is used to find the most common label in a set of labels.<br>
* **counter = Counter(y)** creates a Counter object from the list of labels y. The Counter object is a dictionary-like collection where each unique element in y is a key, and the corresponding value is the count of occurrences of that element in y.
* **value = counter.most_common(1)[0][0]** finds the most common label in y using the most_common method of the Counter object. The most_common method returns a list of tuples, where each tuple contains an element and its count, sorted in descending order of count. By specifying 1 as the argument to most_common, we get the most common element along with its count. Finally, [0][0] is used to extract the most common element from the list of tuples.

The **`_traverse_tree** method is used to traverse the decision tree to predict the label for a given input sample x.<br>
* **if node.is_leaf_node()** This condition checks if the current node is a leaf node. If it is, the method returns the value of the leaf node, which is the predicted label for the input sample x.
* **if x[node.feature] <= node.threshold**: This condition checks if the feature value of the input sample x for the feature corresponding to the current node is less than or equal to the threshold value of the current node. If it is, the method recursively calls _traverse_tree on the left child node of the current node. This means that the traversal moves to the left child node if the feature value is less than or equal to the threshold.
* Then we recursively call this function on the elft child node of the current node if the feature value is less than or equal to the threshold.

## Decision Tree for Regression Task

```python
class DecisionTreeRegressor:
    def __init__(self, min_samples_split=2, max_depth=2):
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        self.root = None

    def build_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        if n_samples >= self.min_samples_split and depth <= self.max_depth:
            best_feature, best_threshold = self.get_best_split(X, y)
            if best_feature is not None and best_threshold is not None:
                left_idxs, right_idxs = self.split_dataset(X, best_feature, best_threshold)
                left_subtree = self.build_tree(X[left_idxs], y[left_idxs], depth + 1)
                right_subtree = self.build_tree(X[right_idxs], y[right_idxs], depth + 1)
                return Node(feature=best_feature, threshold=best_threshold, left=left_subtree, right=right_subtree)
        leaf_value = np.mean(y)
        return Node(value=leaf_value)

    def get_best_split(self, X, y):
        n_samples, n_features = X.shape
        best_feature, best_threshold, best_variance = None, None, float("inf")
        for feature_idx in range(n_features):
            thresholds = np.unique(X[:, feature_idx])
            for threshold in thresholds:
                left_idxs, right_idxs = self.split_dataset(X, feature_idx, threshold)
                if len(left_idxs) > 0 and len(right_idxs) > 0:
                    y_left, y_right = y[left_idxs], y[right_idxs]
                    variance = self.calculate_variance_reduction(y, y_left, y_right)
                    if variance < best_variance:
                        best_feature, best_threshold, best_variance = feature_idx, threshold, variance
        return best_feature, best_threshold

    def split_dataset(self, X, feature_idx, threshold):
        left_idxs = np.where(X[:, feature_idx] <= threshold)[0]
        right_idxs = np.where(X[:, feature_idx] > threshold)[0]
        return left_idxs, right_idxs

    def calculate_variance_reduction(self, y, y_left, y_right):
        variance_parent = np.var(y)
        variance_left = np.var(y_left) if len(y_left) > 0 else 0
        variance_right = np.var(y_right) if len(y_right) > 0 else 0
        n = len(y)
        n_left, n_right = len(y_left), len(y_right)
        variance_reduction = variance_parent - ((n_left / n) * variance_left + (n_right / n) * variance_right)
        return variance_reduction

    def fit(self, X, y):
        self.root = self.build_tree(X, y)

    def predict(self, X):
        return np.array([self._predict_tree(x, self.root) for x in X])

    def _predict_tree(self, x, node):
        if node.is_leaf_node():
            return node.value
        if x[node.feature] <= node.threshold:
            return self._predict_tree(x, node.left)
        return self._predict_tree(x, node.right)
```

This class represents a **Decision Tree** model.<br>
The **`__init__()`** is a constructor method for the class which initializes the attributes of the decision tree model.<br>
* **min_samples_split** attribute specifies the minimum number of samples required to split an internal node. If a node has fewer samples than min_samples_split, it will not be split, and it will become a leaf node.
* **max_depth** attribute specifies the maximum depth of the decision tree. If set to None, the tree will grow until all leaves are pure or until all leaves contain less than min_samples_split samples.
* **root** attribute stores the root node of the decision tree. Initially, it is set to None because the tree is empty.

The **`build_tree()`** method recursively builds the decision tree using the input data X (features) and y (target). It stops splitting nodes when either **min_sample_split** is reached or **max_depth** is exceed. It returns a **Node** object whihc represents a node in the decision tree.<br>

The **`get_best_split()`** method finds the best feature and threshold to split the data X based on the target y. It calculates the variance reduction for each feature and threshold combination and returns the feature index and threshold that result in the maximum variance reduction.<br>

The **`split_dataset()`** method splits the dataset X into two subsets based on a given feature index and threshold. It returns the indices of the samples that fall on the left and right sides of the split.<br>

**`calculate_variance_reduction()`** method calculates the variance reduction, which is used as the splitting criterion. It computes the variance of the target y and the variance of the targets in the left and right subsets. It returns the reduction in variance achieved by the split.<br>

The **`fit()`** method fits the decision tree regressor to the input data X and target y by building the tree using the build_tree method.<br>

The **`predict()`** method predicts the target values for a new set of input features X by traversing the decision tree and returning the average target value of the leaf node reached by each input sample.<br>

The **`_predict_tree()`** method is a helper function for the predict method. It recursively traverses the decision tree starting from the root node node and returns the predicted target value for a single input sample x.<br>
