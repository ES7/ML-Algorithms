# Support Vector Machine (SVM)
SVM falls under the **supervised learning** category, used for classification and regression analysis. Suppose we have labeled training data consisting of triangles and circles. To classify or separate them, we can simply draw a line between these two classes. This line is called a **decision boundary** or **hyperplane** in general terms. The line determines whether new data belongs to the triangle or circle class.<br>
<img src="Images/SVM1.JPEG" width="400" height="150"><br>
In both classes, there exists a data point nearest to the hyperplane. In the given graph, the point A belonging to the circle class is nearest to the hyperplane, and similarly, point B from the triangle class is nearest. After identifying these two points, we draw lines passing through them and **parallel** to the hyperplane. Line L1 passes through point A, and line L2 passes through point B. Lines L1, L2, and the hyperplane are parallel to each other. The distance between line L1 and the hyperplane is D1, and the distance between line L2 and the hyperplane is D2. The sum of D1 and D2 gives the **gap** or **margin**, which is the distance between lines L1 and L2. The margin is significant in deciding the hyperplane. Whether a hyperplane exists or not is determined by the margin. The two points, point A and point B, are called **support vectors**. Line L1 is also called **Negative Hyperplane** and line L2 is called **Positive Hyperplane**.<br>
<img src="Images/SVM2.png" width="400" height="150"><br>
For **linearly separable data**, the data can be separated into two classes by drawing a line between them. In the graph, there are three lines that separate the two classes of circles and squares. How do we decide which hyperplane is best for our data? We choose the hyperplane that generates the **maximum width margin**. Observe that the margin width of **model 3** is minimum, while the margin of **model 2** is maximum. Hence, model 2 is the best fit for this data. Future predictions will be better if the margin width is greater.

## Types of SVM
**1. Linear SVM:** When the data is perfectly linearly separable only then we can use Linear SVM. Perfectly linearly separable means that the data points can be classified into 2 classes by using a single straight line(if 2D).<br>
**2. Non-Linear SVM:** When the data is not linearly separable then we can use Non-Linear SVM, which means when the data points cannot be separated into 2 classes by using a straight line (if 2D) then we use some advanced techniques like kernel tricks to classify them. In most real-world applications we do not find linearly separable datapoints hence we use kernel trick to solve them.

### How to Choose the Right Kernel? 
Selecting an appropriate kernel function is crucial as the model's effectiveness hinges on this choice. The selection of a kernel is entirely dependent on the dataset under consideration. In cases where the dataset is linearly separable, it is advisable to opt for a **linear kernel function**. This choice is favored due to its simplicity and lower complexity compared to other kernel functions.<br>
In scenarios where the data is not linearly separable, more intricate kernels are preferred. Typically, SVM is employed with **RBF and linear kernel** functions, as other options like the polynomial kernel are seldom utilized due to their **lower efficiency**.<br>
<img src="Images/SVM29.png" width="300" height="100"><img src="Images/SVM30.png" width="300" height="100"><br>

Both kernels yield comparable results, performing well with our dataset. However, the choice between them requires consideration. Linear SVM is a parametric model, meaning it describes a model where all data is represented within its parameters. In essence, only the parameters are necessary to predict future values from the current ones.<br>
The complexity of the RBF kernel increases with the size of the training data. Not only is it more computationally expensive to prepare the RBF kernel, but we also need to retain the kernel matrix. Furthermore, projecting into the "infinite" higher-dimensional space where the data becomes linearly separable is more computationally intensive during prediction. If the dataset is not linear, using a linear kernel would lead to significantly lower accuracy.<br>
Thus, for such datasets, opting for the RBF kernel is advisable without hesitation, as it creates a decision boundary that separates the classes effectively.<br>


# Code from Scratch of Linear SVM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from matplotlib.colors import ListedColormap
```

## Generating the Data for Classification

```python
X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
```

We have created a Linearly Separable Data with two classes **'blue'** and **'green'**, now we will make the SVM model that can fit a line to separate the data.

## SVM Model for Linearly Separable Data

```python
class SVM_Model:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        y_ = np.where(y <= 0, -1, 1)
        
        self.w = np.zeros(n_features)
        self.b = 0
        
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.lr * y_[idx]

    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)
```

**1. __init__():** This is the constructor method for the SVM class. It initializes the SVM with hyperparameters including the **learning rate, regularization parameter (lambda_param),** and **number of iterations (n_iters)**. It also initializes the **weights (w) and bias (b)** to **None**.<br>
**2. fit():** This method is used to train the SVM model. It takes input features X and corresponding labels y. It first converts the labels y to **-1 or 1** using **np.where()** to make the problem **binary**. Then, it initializes the weights w and bias b to 0. It then iterates over the dataset **n_iters times** and updates the weights and bias using the SVM **optimization algorithm**. The algorithm tries to **maximize the margin** between the two classes while **minimizing the misclassification error**.<br>
**3. predict():** This method is used to make predictions on new data. It takes input features X and computes the decision function **`np.dot(X, self.w) - self.b`** to determine the predicted class. If the result is greater than 0, it predicts class 1, otherwise it predicts class -1.

## Applying our Model on the Generated Data

```python
# Instantiate the SVM_Model
svm_model = SVM_Model(learning_rate=0.001, lambda_param=0.01, n_iters=1000)

# Fit the model to the data
svm_model.fit(X, y)

# Plot the decision boundary
x1 = np.linspace(-1, 4, 100)
x2 = -(svm_model.w[0] * x1 + svm_model.b) / svm_model.w[1]
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
plt.plot(x1, x2, color='black', linestyle='-', label='Decision Boundary')

# Highlight the support vectors
plt.scatter(X[svm_model.predict(X) == 1][:, 0], X[svm_model.predict(X) == 1][:, 1], c='r', marker='o', label='Support Vectors (Class 1)')
plt.scatter(X[svm_model.predict(X) == -1][:, 0], X[svm_model.predict(X) == -1][:, 1], c='y', marker='o', label='Support Vectors (Class -1)')

plt.legend()
plt.title('SVM Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

The yellow dots are predictions of class -1 and red dots are predictions of class 1.

### Accuracy of Our Model

```python
y_pred = svm_model.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
zero_one_colourmap = ListedColormap(('blue', 'red'))
def plot_decision_boundary(X, y, clf):
    X_set, y_set = X, y
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                                 stop = X_set[:, 0].max() + 1, 
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1, 
                                 stop = X_set[:, 1].max() + 1, 
                                 step = 0.01))
  
    plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), 
                                             X2.ravel()]).T).reshape(X1.shape),
               alpha = 0.75, 
               cmap = zero_one_colourmap)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = (zero_one_colourmap)(i), label = j)
    plt.title('SVM Decision Boundary')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    return plt.show()
```

```python
plot_decision_boundary(X, y, classifier)
```

# SVM for Regression Task
Support Vector Machines (SVMs) can also be used for regression tasks, a variant known as Support Vector Regression (SVR). The goal of SVR is to approximate the mapping function from input variables to the continuous output variable, while also controlling the model complexity and error. Here's a brief explanation of how SVR works for regression tasks. Similar to classification just we need to find the target variable y for different values of X.<br>
$y = W^{T}.X - b$.<br>
The Loss Function will be the difference between the the predicted value and the actual value.<br>
$L(y, y_{pred}) = max(0, |y-y_{pred}| - e)$<br>
here e is the regularization term.

```python
class SVR_Model:
    def __init__(self, epsilon=0.1, learning_rate=0.01, lambda_param=0.01, n_iters=1000):
        self.epsilon = epsilon
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.w = np.zeros(n_features)
        self.b = 0
        
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                y_pred = np.dot(x_i, self.w) + self.b
                error = y[idx] - y_pred

                if error < -self.epsilon:
                    self.w += self.lr * (2 * self.lambda_param * self.w - x_i)
                    self.b += self.lr * 2 * self.lambda_param
                elif error > self.epsilon:
                    self.w += self.lr * (2 * self.lambda_param * self.w + x_i)
                    self.b -= self.lr * 2 * self.lambda_param
                else:
                    self.w += self.lr * 2 * self.lambda_param * self.w

    def predict(self, X):
        return np.dot(X, self.w) + self.b
```

**__init__():** This is the constructor method for the SVR class. It initializes the SVR with hyperparameters including the epsilon (epsilon-insensitive loss threshold), learning rate (learning_rate), regularization parameter (lambda_param), and number of iterations (n_iters). It also initializes the weights (w) and bias (b) to None.<br>
**fit():** This method is used to train the SVR model. It takes input features X and corresponding target values y. It initializes the weights w to zeros and the bias b to 0. Then, it iterates over the dataset n_iters times and updates the weights and bias using the SVR optimization algorithm. The algorithm adjusts the weights and bias based on the error between the predicted and actual target values, ensuring that the error is within the epsilon-insensitive loss threshold.<br>
**predict():** This method is used to make predictions on new data. It takes input features X and computes the predicted target values using the learned weights w and bias b.

## Generating the Data with Noise for SVR

```python
# Generate the data
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel()

# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(20))

#plot
plt.scatter(X, y, color='black', label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Data with Noise')
plt.legend()
plt.show()
```

## Applying our Model on the data

```python
# Fit SVR model
svr = SVR_Model()
svr.fit(X, y)
y_pred = svr.predict(X)

# Plot the results
plt.scatter(X, y, color='black', label='Data')
plt.plot(X, y_pred, color='blue', linewidth=3, label='SVR Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('SVR for Regression')
plt.legend()
plt.show()
```
